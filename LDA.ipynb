```python
import re
import numpy as np

def clean_str(text):
    if text == np.nan:
        return ''
    pattern = '[^\w\s]'       
    text = re.sub(pattern=pattern, repl=' ', string=text)
    pattern = '\n'         
    text = re.sub(pattern=pattern, repl=' ', string=text)
    return text 

df1['text']=df1['프로그램 운영목적 내용'].apply(clean_str)
```

```python
from konlpy.tag import Kkma
kkma=Kkma()
```

```python
def to_nouns(text):
    text_pos=kkma.pos(text)
    nouns=[]

    for a,b in text_pos:
        if len(a)>1:
            if'NN' in b:
                nouns.append(a)
    return nouns
```
