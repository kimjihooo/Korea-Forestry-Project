```python
import pandas as pd

#1. 도시 공원 생활권 데이터 로딩
df1 = pd.read_csv("./data/1_서울시 아파트별 도시공원 생활권 정보_mod.csv", encoding="cp949")
df1.head()
```

```python
df1_group2 = df1[['단지명', '법정동주소', '공원수']]
df1_group2.head()
```

```python
#2. 산책 생활권 정보 로딩
df2 = pd.read_csv("./data/2_서울시 아파트별 산책 생활권 정보_mod.csv", encoding="cp949")
df2.head()
```

```python
df2_group2 = df2[['단지명','코스수']]
df2_group2 = df2_group2.rename(columns={'코스수':'산책 코스수'})
df2_group2.head()
```

```python
a= pd.merge(df1_group2, df2_group2, how='left')
a
```

```python
#3. 한강, 지천 생활권 데이터 로딩
df3 = pd.read_csv("./data/3_서울시 아파트별 한강 및 지천 생활권 정보 (3).csv", encoding="cp949")
df3.head()
```

```python
df3_group2 = df3[['단지명','코스수']]
df3_group2 = df3_group2.rename(columns={'코스수':'하천길 코스수'})
df3_group2.head()
```
